```python
import os
import pympi

# Chemin des dossiers d'entrée et de sortie
input_folder = "/home/or-llsh-156-l01/projets/CREAM/data/Haiti-CMU/annotations/MFA-TextGrid/raw/synth"
output_folder = "/home/or-llsh-156-l01/projets/CREAM/data/Haiti-CMU/annotations/MFA-TextGrid/raw/synth_eaf"

# Créer le dossier de sortie s'il n'existe pas
os.makedirs(output_folder, exist_ok=True)

def textgrid_to_eaf(textgrid_path, eaf_path, skipempty=True, pointlength=0.1):
    """
    Convertit un TextGrid en EAF directement via la méthode to_eaf().
    :param textgrid_path: Chemin vers le fichier TextGrid.
    :param eaf_path: Chemin où écrire le EAF résultant.
    :param skipempty: Si True, ignore les intervalles vides.
    :param pointlength: Durée (en secondes) à assigner aux "points" 
                        (TextTier) pour en faire de courts intervalles 
                        dans l'EAF.
    """
    # 1) Charger le TextGrid
    tg = pympi.Praat.TextGrid(textgrid_path)

    # 2) Convertir en EAF 
    eaf = tg.to_eaf(skipempty=skipempty, pointlength=pointlength)

    # 3) Enregistrer l’EAF
    eaf.to_file(eaf_path)

    print(f"Conversion réussie : {textgrid_path} --> {eaf_path}")


# Traitement des fichiers du dossier d'entrée
for file_name in os.listdir(input_folder):
    if file_name.endswith(".TextGrid"):
        input_path = os.path.join(input_folder, file_name)
        output_path = os.path.join(output_folder, os.path.splitext(file_name)[0] + ".eaf")

        try:
            textgrid_to_eaf(input_path, output_path)
        except Exception as e:
            print(f"[Erreur] Problème inattendu avec le fichier {file_name}. Message : {e}")

print("[Info] Conversion terminée.")
```
